# Dimension Reduction (차원 축소)


차원: 수학에서 공간 내에 있는 점 등의 위치를 나타내기 위해 필요한 축의 개수


중요성: 차원이 늘어나면 동일한 데이터를 가지고 있더라도 데이터 공간을 차지하는 비율은 줄어들고 분석에 요구하는 데이터의 건수도 증가하게 된다. 따라서 중복되는 차원을 제거하고 데이터를 더욱 잘 설명할 수 있는 축으로 변경함으로써 차원을 축소하는 기법은 데이터를 다루는데 중요한 역할을 한다.


적용: 만약 핸드폰에서 홍채 인식을 한다는 가정을 했을 때를 예로 들 수 있다. 핸드폰 카메라에서 가져오는 데이터는 사람의 얼굴과 배경 등 여러가지 요소를 가지고 있을 것이다. 하지만 분석하고자 하는 데이터는 눈의 홍채에 대한 데이터만 요구하고 있기 때문에 홍채에 대한 축으로만 데이터를 구성하는 것이 유리할 것이다.

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline

iris = load_iris()
N = 10  # 앞의 10송이만 선택
X = iris.data[:N, :2]  # 꽃받침 길이와 꽃받침 폭만 선택

plt.plot(X.T, 'o:')
plt.xticks(range(4), ["Length", "Width"])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title("Iris")
plt.legend(["Sample {}".format(i + 1) for i in range(N)])
plt.show()
```

```python
plt.figure(figsize=(8, 8))
ax = sns.scatterplot(0, 1, data=pd.DataFrame(X), s=100, color=".2", marker="s")
for i in range(N):
    ax.text(X[i, 0] - 0.05, X[i, 1] + 0.03, "Sample {}".format(i + 1))
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Iris 2-dimension")
plt.axis("equal")
plt.show()
```
